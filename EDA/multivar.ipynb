<h3>4. Distplot in (numerical-Cateogrical)</h3>

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)#blue one
#pahle jp survived nhi kiye unko filter karke ek alag se group kiya using pandas then age k saath map kiya to isne probabilty distribution diya
#its like new data frame me se age nikal liya
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False) #orange one
#using both at same time we can see complete comaparison
#we can see bache ki probabilty ki marne ki kam h blue wala niche h jo marne ki prob bta raha h
```

```python
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age'], label="Not Survived")
sns.kdeplot(titanic[titanic['Survived'] == 1]['Age'], label="Survived")

#distplot ki jagah se kdeplot aaya h
```

<h3>HeatMap (Categorical - categorical) </h3>

```python
titanic.head()
```

```python
#assume hume nikalna h ki har pclass me kitne mar gye kitne jinda h
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(
    pd.crosstab(titanic['Pclass'],titanic['Survived']),annot=True)
#we can clearly see lighter shade showing more number in 3-0 white shades where we can see more number of people die
```

but isse hume samjh nhi aayega ki percentage of people kitna mare ya kon sa class jyada deadly h so we can use groupby 

```python
titanic.groupby('Pclass')['Survived'].mean()*100
#by this pahle groupby ka use karke pclass me teeno cateogary k group alag kiye phir usko survived k saath dekh k mean nikal diya
```

GroupBy ka kaam hota hai:
data ko groups me baant ke unpe calculation karna

```python
titanic.groupby('Sex')['Survived'].mean()
```

```python
titanic.groupby('Embarked')['Survived'].sum()
```

```python
titanic.groupby('Embarked')['Survived'].mean()
```

count()	Total passengers<br>
sum()	Total survived<br>
mean()	Survival rate

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
titanic.groupby('Pclass')['Embarked'].count()
```
